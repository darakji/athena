```python
from ase.io import read
import numpy as np
import torch

from mace.calculators import MACECalculator

# ================================
# INPUTS
# ================================

# Bulk Li structure (same as DFT)
li_bulk_path = "/home/mehuldarak/athena/li_and_llzo_relaxed_seperate/Li_100_slab__LLZO_001_Zr_code93_sto/Li_100_slab__sc_4x4x1__scaled_to_LLZO_001_Zr_code93_sto.cif"

# Single Li atom (for MACE E0)
li_atom_path = "/home/mehuldarak/athena/single_atoms/Li.cif"

# MACE model
model_path = "/home/mehuldarak/athena/mace_fps_training/checkpoints/mace_fps_split17_SaveIT0_256_candidate3.model"

# DFT total energy (Ry)
E_bulk_dft_Ry = -2263.88815952

# DFT atomic reference (eV)
E0_Li_DFT = -190.7590256408

# ================================
# LOAD STRUCTURES
# ================================
li_bulk = read(li_bulk_path)
li_atom = read(li_atom_path)

N = len(li_bulk)

print(f"Number of Li atoms = {N}")

# ================================
# LOAD MACE MODEL
# ================================
device = "cuda" if torch.cuda.is_available() else "cpu"

calc = MACECalculator(
    model_paths=[model_path],
    device=device,
    default_dtype="float64"
)

li_bulk.set_calculator(calc)
li_atom.set_calculator(calc)

# ================================
# MACE ENERGIES
# ================================

# Bulk
E_bulk_mace = li_bulk.get_potential_energy()
E_bulk_mace_per_atom = E_bulk_mace / N

# Single atom → MACE E0
E0_Li_MACE = li_atom.get_potential_energy()

# Cohesive energy (MACE, now consistent)
E_coh_mace = E_bulk_mace_per_atom - E0_Li_MACE

# ================================
# DFT ENERGIES
# ================================
RY_TO_EV = 13.605693

E_bulk_dft = E_bulk_dft_Ry * RY_TO_EV
E_bulk_dft_per_atom = E_bulk_dft / N

# Cohesive energy (DFT)
E_coh_dft = E_bulk_dft_per_atom - E0_Li_DFT

# ================================
# COMPARISON
# ================================
delta_E_per_atom = E_bulk_mace_per_atom - E_bulk_dft_per_atom
delta_coh = E_coh_mace - E_coh_dft

# ================================
# PRINT RESULTS
# ================================
print("\n==============================")
print("ENERGY COMPARISON")
print("==============================")

print(f"MACE total energy       = {E_bulk_mace:.6f} eV")
print(f"DFT total energy        = {E_bulk_dft:.6f} eV")

print("\n--- Per Atom ---")
print(f"MACE energy/atom        = {E_bulk_mace_per_atom:.6f} eV")
print(f"DFT energy/atom         = {E_bulk_dft_per_atom:.6f} eV")

print("\n--- Atomic Reference ---")
print(f"MACE E0 (Li atom)       = {E0_Li_MACE:.6f} eV")
print(f"DFT E0 (Li atom)        = {E0_Li_DFT:.6f} eV")

print("\n--- Cohesive Energy ---")
print(f"MACE cohesive energy    = {E_coh_mace:.6f} eV/atom")
print(f"DFT cohesive energy     = {E_coh_dft:.6f} eV/atom")

print("\n--- Errors (meV/atom) ---")
print(f"Δ Energy/atom           = {1000*delta_E_per_atom:.6f} meV/atom")
print(f"Δ Cohesive energy       = {1000*delta_coh:.6f} meV/atom")

print("==============================")
```
