```python
import numpy as np
from ase import Atoms
from mace.calculators import MACECalculator

# 35 bohr → Å
bohr_to_ang = 0.529177
box = 35 * bohr_to_ang

# create isolated atom structures
elements = ["Li", "La", "Zr", "O"]

atoms_list = []
for el in elements:
    atoms = Atoms(
        symbols=[el],
        positions=[[0, 0, 0]],
        cell=[box, box, box],
        pbc=False
    )
    atoms_list.append(atoms)

# load MACE model
calc = MACECalculator(
    model_paths="/home/mehuldarak/athena/mace_seed_1_trial0.model",
    device="cuda"  # or "cpu"
)

# run inference
for atoms in atoms_list:
    atoms.calc = calc
    energy = atoms.get_potential_energy()
    print(atoms.symbols[0], energy)
```

```python
from ase.io import read
import pandas as pd
from collections import Counter

E0s = {3: -0.29754725, 8: -1.54838784, 40: -2.28579303, 57: -0.45580806}
files = ["/home/mehuldarak/athena/fpsSeed_univOMAT.extxyz"]

rows = []

for f in files:
    atoms_list = read(f, ":")

    for atoms in atoms_list:
        filename = atoms.info["source_file"]
        ref_energy = atoms.info["mace_energy"]

        counts = Counter(atoms.get_atomic_numbers())
        e0_sum = sum(E0s[z] * n for z, n in counts.items())

        cohesive_energy = ref_energy - e0_sum

        rows.append({
            "Filename": filename,
            "REF_energy": ref_energy,
            "Cohesive_energy": cohesive_energy
        })

df = pd.DataFrame(rows)

df.to_csv("universalOnFPScohesive_energies.csv", index=False)

print("Saved to cohesive_energies.csv")
```
